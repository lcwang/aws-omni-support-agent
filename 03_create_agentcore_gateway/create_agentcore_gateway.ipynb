## 把第二步中创建的aws support tool lambda添加到Agentcore gateway的target中

## store agentcore gateway url in parameter store

```python
import boto3
session = boto3.session.Session()
region =  session.region_name
```

## replace the gateway url

```python
param_customer_support_gateway = "/support/agentgateway/aws_support_gateway"
gatewa_url = "https://gateway-support-mcp-xxxxxx.gateway.bedrock-agentcore.us-east-1.amazonaws.com/mcp"
```

```python
import boto3
ssm = boto3.client("ssm",region_name=region)
response = ssm.put_parameter(
    Name=param_customer_support_gateway,
    Value=gatewa_url,
    Type="String",
    Overwrite=True  # True = 已存在就更新
)
print("Parameter stored:", response)
```
